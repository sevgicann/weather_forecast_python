Extract year and month:

```python
df_w["date"] = pd.to_datetime(df_w["date"], format = '%Y-%m-%d')
df_w['year'] = df_w['date'].dt.year
df_w["month"] = df_w["date"].dt.month
print(df_w.head())
```

Analyze to years:

```python

plt.figure(figsize=(15, 10))
plt.title("Delhi'de Yıllara Göre Ortalama Sıcaklık Değişimi", fontsize=16)

sns.lineplot(data=df_w, x='month', y='meantemp', hue='year', marker='o', markersize=8, linewidth=2)
plt.xlabel('Ay', fontsize=14)
plt.ylabel('Ortalama Sıcaklık (°C)', fontsize=14)

sns.set_palette("husl")
plt.legend(title='Yıl', title_fontsize='14', fontsize='12', loc='upper right')
sns.despine()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

We can see a rise in the average temperature of Delhi every year.

Here, prophet library is a library developed by Facebook that provides a forecasting model used specifically for time series data. This library requires a specific format: the ds column contains time information (usually date) and the y column contains the target variable to be predicted.


```python
forecast_data = df_w.rename(columns = {"date": "ds", 
                                       "meantemp": "y"})
print(forecast_data)
```

```python
import plotly.express as px
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
```

```python
import plotly.express as px
fig = plot_plotly(model, predictions)

# Özelleştirmeleri ekleyelim
fig.update_layout(
    title="Prophet Tahmin Grafiği",
    xaxis_title="Tarih",
    yaxis_title="Meantemp",
    template="plotly_dark",  # İsteğe bağlı: Farklı bir şablon seçebilirsiniz
)

# Eksen etiketleri ve büyüklüğü gibi özellikleri güncelleyelim
fig.update_xaxes(tickangle=45, tickformat="%Y-%m-%d", title_text="Tarih")
fig.update_yaxes(title_text="Meantemp")

# Grafik öğelerini daha da özelleştirebilirsiniz
# Örneğin, çizgi genişliği, renk, marker stil, vs.
fig.update_traces(line=dict(width=2, color='orange'), marker=dict(size=8, symbol='circle'))

fig.show()
```
